# Programme QUACK (Quasi-Unlimited Algorithm for Creating Kanard)

```python
import os
import random
from tqdm import tqdm
```

```python
latex_pattern = lambda x: rf"""\documentclass{{standalone}}
\usepackage{{tikzducks}}
\begin{{document}}
\begin{{tikzpicture}}
	\duck[
		{x}
	]
\end{{tikzpicture}}
\end{{document}}"""
```

```python
attribut = [
    {"body": ["blue", "red", "yellow"]},
    {"bill": ["cyan", "orange", "green"]},
    "grumpy",
    "laughing",
    "parrot",
    "longhair"
]
```

```python
os.chdir("..")
```

```python
def create_duck_pdf(duck_number, attribut, attribut_number):
    for number in range(duck_number):
        duck_cosmetics = ""
        for key in random.sample(attribut, attribut_number):
            if isinstance(key, dict):
                key, options = list(key.keys())[0], list(key.values())[0]
            else:
                options = []
            duck_cosmetics += f",{key}"
            if len(options) > 0:
                duck_cosmetics += f"={random.choice(options)}"
        duck_cosmetics = duck_cosmetics[1:]
        with open(f"dataset/{number}.tex", "w+") as file:
            file.write(latex_pattern(duck_cosmetics))
```

```python
create_duck_pdf(100, attribut, 2)
```
